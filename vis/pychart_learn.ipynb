## 多种图重叠

```python
# 虚假数据
x_data = ['Apple', 'Huawei', 'Xiaomi', 'Oppo', 'Vivo', 'Meizu']
y_data_bar = [123, 153, 89, 107, 98, 23]
y_data_line = [153, 107, 23, 89, 123, 107]


bar = (Bar()
       .add_xaxis(x_data)
       .add_yaxis('', y_data_bar)
       )

line = (Line()
        .add_xaxis(x_data)
        .add_yaxis('', y_data_line)
        )

overlap = bar.overlap(line)
overlap.render_notebook()
```
